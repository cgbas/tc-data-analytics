```python
import pandas as pd
import os


user = os.getlogin()
exportação = pd.read_csv(f'./ExpVinho.csv', sep=';')

df_transposto = exportação.melt(id_vars=['Id', 'País'], var_name='Ano', value_name='Valor')
df_transposto[['Ano', 'Tipo']] = df_transposto['Ano'].str.split('.', expand=True)
df_transposto['Tipo'] = df_transposto['Tipo'].map({None : 'Quantidade', '1': 'Valor'})
df_pivot = df_transposto.pivot_table(index=['Id', 'País', 'Ano'], columns='Tipo', values='Valor')
exportação_tratado = df_pivot.reset_index()
exportação_tratado.columns.names = [None]
pais_origem = 'Brasil'
exportação_tratado['Pais_origem'] = pais_origem
exportação_tratado = exportação_tratado[['Ano', 'Pais_origem', 'País', 'Quantidade', 'Valor']]
display(exportação_tratado)
display(exportação_tratado.dtypes)
```
